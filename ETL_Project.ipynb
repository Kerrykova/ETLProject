## Extract the CSVs into DataFrames

```python
aac_outcomes_csv = "Resources/Austin_Animal_Center_Outcomes.csv"
aac_outcomes_df = pd.read_csv(aac_outcomes_csv)
aac_outcomes_df.head()
```

```python
city_satisfaction_csv = "Resources/Community_Survey_2015__2016___2017.csv"
city_satisfaction_df = pd.read_csv(city_satisfaction_csv)
city_satisfaction_df.head()
```

## Transform the DataFrames

```python
new_aac_outcomes_df = aac_outcomes_df[['DateTime', 'Outcome Type', 'Animal Type', 'Breed', 'Age upon Outcome']].copy()
new_aac_outcomes_df = new_aac_outcomes_df.rename(index=str, columns={"DateTime": "year", "Outcome Type": "outcome_type", "Animal Type": "animal_type", "Breed":"breed", "Age upon Outcome": "age_upon_outcome"})
new_aac_outcomes_df['year']
# new_animals_df.drop_duplicates("id", inplace=True)
```

```python
new_aac_outcomes_df['year'] = pd.DatetimeIndex(new_aac_outcomes_df['year']).year
```

```python
new_aac_outcomes_df['year'].value_counts()
#we need to drop 2013, 2014, 2018, 2019
```

```python
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2013]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2014]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2018]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2019]
```

```python
new_aac_outcomes_df['outcome_type'] = new_aac_outcomes_df['outcome_type'].replace('Disposal', 'Died')
```

```python
new_aac_outcomes_df['outcome_type'] = new_aac_outcomes_df['outcome_type'].replace('Rto-Adopt', 'Return to Owner')
```

```python
new_aac_outcomes_df.head()
```

```python
new_aac_outcomes_df.groupby(['year','animal_type','outcome_type']).count()
```

```python
new_aac_outcomes_df.head()
```

```python
new_city_satisfaction_df = city_satisfaction_df[["ID", "Year", "Overall quality of life in the city", "Overall quality of city parks and recreation", "Animal services (shelter, adoptions, animal control, etc.)","I feel safe in my neighborhood during the day", "I feel safe in my neighborhood at night","I feel safe in city parks","Number of City of Austin parks","Number of City of Austin walking/biking trails","Appearance of City of Austin park grounds located in Austin","Which of the following best describes your AGE?","Which of the following best describes your annual household income?","Race/Ethnicity Recode"]].copy()
new_city_satisfaction_df = new_city_satisfaction_df.rename(index=str, columns={"ID": "id", "Year": "year", "Overall quality of life in the city": "quality_of_life", "Overall quality of city parks and recreation":"parks_and_rec", "Animal services (shelter, adoptions, animal control, etc.)":"animal_services","I feel safe in my neighborhood during the day":"safety_day","I feel safe in my neighborhood at night":"safety_night","I feel safe in city parks":"safety_parks","Number of City of Austin parks":"num_parks","Number of City of Austin walking/biking trails":"num_trails","Appearance of City of Austin park grounds located in Austin":"park_appearance","Which of the following best describes your AGE?":"age_range","Which of the following best describes your annual household income?":"income_range","Race/Ethnicity Recode":"race_ethnicity"})
# new_city_satisfaction_df.drop_duplicates("id", inplace=True)
new_city_satisfaction_df.set_index("id", inplace=True)
new_city_satisfaction_df.head()
```

```python
new_city_satisfaction_df.groupby(['year']).count()
```

## Create Connection

```python
rds_connection_string = password
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

## Load DataFrames into Database

```python
new_aac_outcomes_df.to_sql(name='aac_outcomes', con=engine, if_exists='append', index=False)
```

```python
new_city_satisfaction_df.to_sql(name='city_satisfaction', con=engine, if_exists='append', index=True)
```
